## Numpy reshape

numpy의 shape을 변경할 수 있음.  
이때, element의 개수는 반드시 동일해야 함

```python
arr = np.arange(1, 13, 1)
```

```python
arr
```

```python
arr34 = arr.reshape(3, 4)  # reshape performance cost is so cheap.
#  Why?
#  넘파이는 2가지로 나눠져 버퍼메모리와 뷰
#  버퍼메모리에는 연속적인 배열이 저장되어있어
#  뷰에는 shape, axis, dim 이런것들이 저장되어 있어
#  shape, axis, dim은 사용자에게 알려주기 위한 것이 아니라
#  버퍼를 어떻게 읽을지에 대해서 지정해둔거야
#  dim은 index
#  shape은 각 index당 가질 수 있는 element의 개수
#  let shape is (3,4)
#  shape, dim, axis 숫자만 바꾸는거라서 엄청 cost가 cheap
arr34
```

```python
arr34[2,3] = 100
arr34
```

```python
arr # 주의! 원래 arr도 바뀌어
#  same buffer but differenct view
#  we just made arr34's view. not copied arr's buffer memory.
```

#### reshape -1의 활용
-1은 해당 axis를 numpy가 자동으로 판단하게 함

```python
arr.reshape(4, -1)
```

reshape(-1, 1)은 데이터 형태를 맞추기 위하여 자주 활용됨

```python
arr34.reshape(-1,1)
```

reshape(1, -1) == flatten()

```python
arr34.flatten()
```

## Aggregate function

numpy는 배열의 합, 평균, 최대/최소값등의 함수를 제공함.
이 함수들은 axis를 기준으로 계산됨

```python
arr = np.arange(0, 32).reshape([4,2,4])
```

```python
print(arr.shape)
arr
```

```python
arr.sum()
```

```python
arr.sum(axis=0)
# 괄호를 기준으로 axis를 생각해라
# 첫번째 괄호 무시하고 가장 바깥쪽 괄호가 axis 0
# 두번째 괄호가 axis 1
# 세번째 괄호가 axis 2
```

```python
arr.sum(axis=(1, 2))
```

```python
arr.sum(axis=1)
```

```python
arr.sum(axis=2)
```

배열의 평균, min, max도 비슷하게 동작함

```python
arr.mean(), arr.min(), arr.max()
```

```python
print(arr.mean(axis=0))
print(arr.min(axis=1))
print(arr.max(axis=2))
```
