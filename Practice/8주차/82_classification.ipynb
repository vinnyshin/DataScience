```python
import pandas as pd
import os
import numpy as np
```

```python
file = 'mobile_price.csv'
data = pd.read_csv(os.path.join(basicpath, file))
```

```python
data.head()
```

## Data preprocessing

### Train_test split

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

## Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
md_random_forest = RandomForestClassifier()
```

```python
md_random_forest.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
matrix = confusion_matrix(y_test, md_random_forest.predict(X_test))
matrix
```

## K-fold cross validation

### 단순 score 계산

sklearn.model_selection.cross_val_score(estimator, X, y=None,...)

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(md_random_forest, X, y, cv=5)
```

```python
score
```

### Iterators KFold 활용

```python
from sklearn.model_selection import KFold
```

```python
X = X.values
y = y.values
kfold = KFold(n_splits=5)
for train_index, test_index in kfold.split(X):
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]

  md_random_forest.fit(X_train, y_train)
  matrix = confusion_matrix(y_test, md_random_forest.predict(X_test))
  print(matrix)
```

## SVM 모델

```python
from sklearn.svm import SVC
```

```python
file = 'mushrooms.csv'
data = pd.read_csv(os.path.join(basicpath, file))
```

```python
data
```

```python
 from sklearn.preprocessing import LabelEncoder
 labelencoder = LabelEncoder()
```

```python
for col in data.columns:
  data[col] = labelencoder.fit_transform(data[col])
```

```python
X = data.iloc[:, 1:]
y = data.iloc[:, 0] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

## Model Test

```python
predict = model.predict(X_test) # 모델 predict값 생성
```

## 정확도 판단

### Confution matrix

```python
matrix = confusion_matrix(y_test, model.predict(X_test))
matrix
```

### Precision, recall

```python
from sklearn.metrics import precision_score, recall_score
```

```python
print(precision_score(y_test, predict, average=None))
print('average: ', precision_score(y_test, predict, average='weighted'))
```

```python
print(recall_score(y_test, predict, average=None))
print('average: ', recall_score(y_test, predict, average='weighted'))
```
