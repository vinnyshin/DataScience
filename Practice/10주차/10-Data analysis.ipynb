## Matplotlib 기본 활용법

Matplotlib: 다양한 그래프 및 다이어그램을 그리는 도구를 지원하는 라이브러리

### 기본적인 그래프

- plt.figure(num=None, figsize=None, ...) - 새로운 figure 생성
- plt.plot(* args, ..., ** kwargs) / args: [x], y, [fmt] - x에 대한 y의 선형 그래프. 이때 x와 y의 크기가 같아야함
- plt.show() - 모든 figure를 display

```python
x = np.arange(50)
y = np.arange(50)
```

```python
x, y
```

#### 기본 ploting

```python
plt.figure(figsize=(4,4))
# plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.show()
```

#### 여러 figure ploting

```python
y2 = np.arange(50, 100)
```

```python

plt.figure(figsize=[5, 5])
plt.plot(x, y)
plt.plot(x, y2)
plt.show()

plt.figure(figsize=[5, 5])
plt.plot(x, y2)
plt.show()
```

#### Subplot의 활용

```python
# 하나의 캔버스 안에 4개의 subplot을 그릴 수 있다.
fig, axes = plt.subplots(2,2, figsize=[8, 8])
axes[0, 0].plot(x, y)
axes[0, 1].plot(x, y*y)
axes[1, 0].plot(x, y**3)
axes[1, 1].plot(x, -y)
plt.show()
```

#### Plot 옵션

```python
fig, axes = plt.subplots(2,2, figsize=[8, 8])
axes[0, 0].plot(x, y)        # plot x and y using default line style and color

# 산포도라고 얘기하는 것, 여기서는 점들끼리 연결되지 않아
# bo는 blue and 'o' 라고 얘기하는 것
# 파란색 동그라미로 점들을 찍어라
axes[0, 1].plot(x, y, 'bo')  # plot x and y using blue circle markers

axes[1, 0].plot(y)           # plot y using x as index array 0..N-1

axes[1, 1].plot(y, 'r+')     # ditto, but with red plusses
plt.show()
```

```python
data.corr()
```

#### Draw Heatmap

```python
corr = data.corr()
fig = plt.figure(figsize=(12,12))
r = sns.heatmap(corr, cmap='Greens', annot=True)

# corr가 높은 데이터를 제거한다. linear regression을 할 때 종속성이 영향을 미친다.
# 그러면 corr가 높은거 다 없애면 linear regression이 항상 잘 되냐?
# 항상 그런것도 아니다. 그러면 corr가 낮다면 데이터가 독립이냐? 그것도 아니다.
# Y = X^2같은것에 대해서 corr를 재면 corr가 0.05이렇게 나오는데 독립은 아니야 종속이지
```

참고 - color map: https://matplotlib.org/3.2.1/tutorials/colors/colormaps.html

```python
from pandas.plotting import scatter_matrix

attributes = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Speed']
# corr가 높은 것들 5개만 뽑아온 것
scatter_matrix(data[attributes], figsize=(16, 16))
```

### Scatter plot

```python
x = data['Attack']
y = data['Defense']
plt.scatter(x, y)
# plot r+, 내지는 bo로 scatter plot찍을 수도 있어
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), c='r')
# 실제 세계에서 0.4정도의 corr를 갖는다고 하면 높은 corr를 가지는 것
# 0.3 넘으면 어느정도 있다..
```

```python
plt.figure(figsize=[15, 10])
sns.boxplot(data = data, x = 'Type 1', y = 'Attack', hue = 'Legendary', palette='Set2')
# RGBA 시스템은 한계가 있어, 그래서 컴퓨터에서 표현하기 위한 HSL이라는 새로운 색상 체계를 만들었어
# HSL에서의 H가 Hue야, 색조를 의미
# S는 Saturation 채도
# L은 Lightness 명도
# 그러면 이 Hue가 왜 categorical data를 나타내는데 활용이 되는가?
# 0과 1의 class 가 있다면 Hue의 0도부터 360도 사이의 2개로 나눠서 보여줘
```

### Quantile plot

```python
from scipy.stats import probplot
```

```python
fig, axes = plt.subplots(2,2, figsize=[10, 10])

# x축이 f value가 되고 y축이 실제 value
# f value는 정규분포를 따르는 데이터를 의미
# 빨간선과 데이터가 많이 안멀수록 정규분포와 비슷하다고 의미
probplot(data['Speed'], plot=axes[0, 0])
axes[0, 0].set_title('Speed')
probplot(data['Attack'], plot=axes[0, 1])
axes[0, 1].set_title('Attack')
probplot(data['Defense'], plot=axes[1, 0])
axes[1, 0].set_title('Defense')
plt.show()

# 실생활에서 많이 쓰이지 않아
```

### Line plot

```python
train = pd.read_csv(basicpath + 'biketrain.csv', parse_dates=['datetime'])
# parse_dates = datetime이라고 하면 문자열의 날짜 데이터를 가지고 파이썬의 datetime 형식으로 바꿔준다.
```

```python
train.head() # 특정 지점을 잡고 지나가는 자전거가 얼마나 많았는지를 계산
```

```python
train['datetime']
# datetime64 형식을 가지고 있다.
```

```python
train["datetime-year"] = train["datetime"].dt.year 
train["datetime-month"] = train["datetime"].dt.month 
train["datetime-day"] = train["datetime"].dt.day 
train["datetime-hour"] = train["datetime"].dt.hour 
train["datetime-minute"] = train["datetime"].dt.minute 
train["datetime-second"] = train["datetime"].dt.second
```

```python
train.head()
```

matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

```python
# 하루동안의 자전거 대수의 변화 추이를 보고 싶어
plt.figure(figsize=[16, 4])
plt.plot(train[:24]['datetime-hour'], train[:24]['count'])
```
